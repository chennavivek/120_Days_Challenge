# Data Manipulation and Analysis in python  

```python
import pandas as pd
```

```python
df = pd.read_csv('Car_Sales.csv')  # Reading a CSV file into a DataFrame
df.head(5) # Displaying the first 5 rows of the DataFrame
```

```python
df.tail(5)  # Displaying the last 5 rows of the DataFrame
```

```python
df.describe()  # Getting a statistical summary of the DataFrame
```

```python
df.dtypes  # Displaying the data types of each column
```

```python
# Handling missing values
df.isnull().sum()  # Checking for missing values in each column
```

```python
df.isnull().any()
```

```python
# Renaming columns
df=df.rename(columns={'Model':'Car_Model'}) # Renaming the 'Model' column to 'Car_Model'
df.head(5)  # Displaying the first 5 rows after renaming a column
```

```python
# changing the value of data type
df['Price']= df['Price_in_thousands'].fillna(0).astype(int)  # Changing the data type of 'Price_in_thousands' to integer
# fillna(0) # is used to replace NaN values with 0 before changing the data type
df.head(5)  # Displaying the first 5 rows after changing the data type
```

```python
# changing the value of data type
df['Price']= df['Price_in_thousands'].fillna(df['Price_in_thousands'].mean()).astype(int)  # Changing the data type of 'Price_in_thousands' to integer
# fillna(df['Price_in_thousands'].mean()) # is used to replace NaN values with the mean of the column before changing the data type
df.head(5)  # Displaying the first 5 rows after changing the data type
```

```python
# using lambda function to change the value of data type
df['Price'] = df['Price_in_thousands'].apply(lambda x:x*2)
df.head(5)  # Displaying the first 5 rows after applying the lambda function to change the data type
```

```python
# Data Aggregation amd Grouping
df.head(5)  # Displaying the first 5 rows of the DataFrame for context
```

```python
grouped_mean = df.groupby('Manufacturer')['Sales_in_thousands'].mean()  # Grouping by 'Car_Model' and calculating the mean of 'Sales_in_thousands'
print(grouped_mean)  # Displaying the mean sales for each car model
```

```python
grouped_sum=df.groupby(['Horsepower', 'Manufacturer'])['Sales_in_thousands'].sum()
grouped_sum 
```

```python
df.groupby(['Horsepower', 'Manufacturer'])['Sales_in_thousands'].mean()
```

```python
# multiple aggregations
grouped_agg=df.groupby('Manufacturer')['Sales_in_thousands'].agg(['mean', 'sum', 'count'])  # Grouping by 'Horsepower' and applying multiple aggregation functions
print(grouped_agg)  # Displaying the aggregated results
```

```python
# Merging and Joining DataFrames
df1 = pd.DataFrame({'Key':['A', 'B', 'C'], 'Value1':[1, 2, 3]})  # Creating a sample DataFrame df1
df2 = pd.DataFrame({'Key':['A', 'B', 'D'], 'Value2':[4, 5, 6]})  # Creating a sample DataFrame df2
df1.head()  # Displaying the first 5 rows of df1
```

```python
df2.head()  # Displaying the first 5 rows of df2
```

```python
merged_df = pd.merge(df1, df2, on='Key', how='inner')  # Merging df1 and df2 on 'Key' using an inner join
print(merged_df)  # Displaying the merged DataFrame
```

```python
merged_df = pd.merge(df1, df2, on='Key', how='outer')  # Merging df1 and df2 on 'Key' using an outer join
print(merged_df)  # Displaying the merged DataFrame
```

```python
merged_df = pd.merge(df1, df2, on='Key', how='left') # Merging df1 and df2 on 'Key' using an outer join
print(merged_df) 
```

```python
merged_df = pd.merge(df1, df2, on='Key', how='right') # Merging df1 and df2 on 'Key' using an outer join
print(merged_df) 
```
